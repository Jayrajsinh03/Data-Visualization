# MatplotLib 

MatplotLib is a plotting library for the Python programming language and its numerical mathematics extension Numpy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt or GTK+. 

Some of the major pros of MatlplotLib are: 
    
    * Easy to use for simple plots
    * Support for custom labels & texts
    * Great control of every element in a figure
    * High-quality output in many formats
    * Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline

# Using %matplotlib inline is useful as we don't require plt.show() function to display the graph everytime. 
```

```python
import numpy as np
```

```python
## Creating Arrays

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
## Plotting scatter graph using matplotlib

plt.scatter(x,y, c = 'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.jpeg')
```

```python
y = x*x
```

```python
## Plotting the plotline

plt.plot(x,y,'r*-', linestyle = 'dashed', linewidth = 2, markersize = 12)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2D Diagram')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*-')

plt.subplot(2,2,3)
plt.plot(x,y,'bo')
```

```python
x = np.arange(1,11)
y = 4*x - 7
plt.title('Matplotlib')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show()
```

```python
# Computing X & Y coordinates for points on sine curve

x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title('sine wave form')

# Plot the points using matplotlib
plt.plot(x, y)
plt.show()
```

```python
# Subplot()
# Compute X and Y coordinates for points on sine and cosine curves

x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1
# Set the first such subplot as active. 
plt.subplot(2,1,1)

#Make the first plot
plt.plot(x, y_sin, 'r*-')
plt.title('Sine')

#Set the second subplot as active, and make the second plot
plt.subplot(2,1,2)

plt.plot(x, y_cos, 'g--')
plt.title('Cosine')

#Show the figure
plt.show()
```

### Bar Graph

```python
## Bar Plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2, color = 'g')
plt.title('Bar graph')

plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

### Histogram

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("histogram")
plt.show()
```

### Box-Plot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

# Rectungalr box plot
plt.boxplot(data, vert = True, patch_artist = True)
```

### Pie Chart

```python
# Data to plot

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.1, 0.1, 0.1)

# Plot
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = False)

plt.axis('equal')
plt.show()
```
